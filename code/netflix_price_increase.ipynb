```python
import matplotlib.pyplot as plt
import seaborn as sns
# import matplotlib 


dates = [2011, 2013, 2014, 2015, 2017, 2019, 2020, 2022]
values = [7.99, 7.99, 8.99, 9.99, 10.99, 12.99, 13.99, 15.49]

# plt.scatter(dates,values)
# plt.scatter(dates, values)
# sns.set_style('darkgrid')
sns.set_palette("RdBu")
# sns.color_palette("flare", as_cmap=True)
sns.regplot(x=dates, y=values)
# plt.plot(dates, values)
plt.xlabel('year')
plt.ylabel('cost (USD)') 
plt.xticks(range(2011, 2023))
plt.title('Netflix (standard) Time vs Price')
# plt.legend()

plt.savefig('netflix.jpg')
plt.show()
```
